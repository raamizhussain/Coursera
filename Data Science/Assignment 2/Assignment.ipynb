```python
# 1

import yfinance as yf
tesla_df = yf.download('TSLA', start='2010-01-01', end='2023-12-31')
tesla_df.reset_index(inplace=True)
tesla_df.head()
```

```python
# 2

import requests
from bs4 import BeautifulSoup
import pandas as pd

url_tesla = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
html_data = requests.get(url_tesla).text
soup_tesla = BeautifulSoup(html_data, "html.parser")

tables = pd.read_html(str(soup_tesla.find_all("table")))
tesla_revenue = tables[1]
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].replace({'\$':'', ',':''}, regex=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"])
tesla_revenue.head()
```

```python
# 3

gme_df = yf.download('GME', start='2010-01-01', end='2023-12-31')
gme_df.reset_index(inplace=True)
gme_df.head()
```

```python
# 4

url_gme = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_data_gme = requests.get(url_gme).text
soup_gme = BeautifulSoup(html_data_gme, "html.parser")

tables_gme = pd.read_html(str(soup_gme.find_all("table")))
gme_revenue = tables_gme[1]
gme_revenue.columns = ["Date", "Revenue"]
gme_revenue["Revenue"] = gme_revenue["Revenue"].replace({'\$':'', ',':''}, regex=True)
gme_revenue = gme_revenue[gme_revenue['Revenue'] != ""]
gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"])
gme_revenue.head()
```

```python
# 5

import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

ax[0].plot(tesla_df['Date'], tesla_df['Close'], label='Tesla Stock Price', color='tab:blue')
ax[0].set_ylabel('Stock Price (USD)')
ax[0].legend()

ax[1].bar(tesla_revenue['Date'], tesla_revenue['Revenue'], color='tab:green', alpha=0.7)
ax[1].set_ylabel('Revenue (USD)')
ax[1].set_xlabel('Date')

plt.suptitle("Tesla Stock Price vs Revenue", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 6

fig, ax = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

ax[0].plot(gme_df['Date'], gme_df['Close'], label='GameStop Stock Price', color='tab:red')
ax[0].set_ylabel('Stock Price (USD)')
ax[0].legend()

ax[1].bar(gme_revenue['Date'], gme_revenue['Revenue'], color='tab:orange', alpha=0.7)
ax[1].set_ylabel('Revenue (USD)')
ax[1].set_xlabel('Date')

plt.suptitle("GameStop Stock Price vs Revenue", fontsize=16)
plt.tight_layout()
plt.show()
```
